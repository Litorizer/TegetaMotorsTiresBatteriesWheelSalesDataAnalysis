# იანვარი


```python

def plot_cumulative_cases(Cat_name):
    category = January[January['Category'] == Cat_name]
    plt.plot(category['Date'], category['Quantity'])
    plt.title('{}: Cumulative Sales'.format(Cat_name))
    plt.xlabel('Date')
    plt.ylabel('Sales')
    plt.xticks(rotation=45)
    plt.show()
    
plot_cumulative_cases("Oil")

plot_cumulative_cases("Tire")

plot_cumulative_cases("Battery")

```
